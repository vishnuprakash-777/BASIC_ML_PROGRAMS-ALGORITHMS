```python
import pandas as pd
df=pd.read_csv('/home/vishnu/Documents/pythonsem3/customer.csv')
df
```

```python
nominal_features = ['user_type']
df_nominal=pd.get_dummies(df[nominal_features])
```

```python
gen_ord_map = {'Class A':1,'Class B':2,'Class C':3}
```

```python
df['user_class_ordinal']=df['user_class'].map(gen_ord_map)
```

```python
categorical_features = ['user_class','user_type']
df_categorical=pd.get_dummies(df[categorical_features])
```

```python
df_transformed=pd.concat([df,df_nominal,df_categorical],axis=1)
```
